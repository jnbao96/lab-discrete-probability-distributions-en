## Binomial Distribution

Similar to the above problem, this time again we randomly sample 20 fruits.

**What is the probability that this time the sample contains 5 apples and 15 oranges?**

Please note that this time the order of the fruits being picked does not matter.

````python
# your solution here
import scipy.stats as stats
from scipy.stats import binom
````

````python
# your solution here
prob5_apple = binom.pmf(5,20,0.6)
print(f'The probability that this time the sample contains 5 apples and 15 oranges is: {prob5_apple:.3f}')
````

````python
# if we look at the oranges, should be the same value obtained
prob15_orange = binom.pmf(15, 20, 0.4)
print(f'The probability that this time the sample contains 5 apples and 15 oranges is: {prob15_orange:.3f}')
````

In the lesson we also discussed CDF (cumulative distribution function). In the context of the above problem, **what is the probability that less than 5 fruits picked are apples?**

Write the formula and calculate the probability below.

````python
# your code here
# P(x<5) = p(X=0) + p(X=1) + p(X=2) + p(X=3) + p(X=4)
P0 = 1 * (0.6)**0 * (0.4)**20
P1 = (20/1) * (0.6)**1 * (0.4)**19
P2 = (20*19 / 2) * (0.6)**2 * (0.4)**18
P3 = (20*19 *18 / (3*2*1)) * (0.6)**3 * (0.4)**17
P4 = (20*19 *18 *17 / (4* 3*2*1)) * (0.6)**4 * (0.4)**16

cdf_less5 = P0 + P1 + P2 + P3 + P4

print(f'the probability that less than 5 fruits picked are apples: {cdf_less5:3f}')
````

## Poisson Distribution

In this challenge you are required to use the Math library in python to solve the problems.

In the UEFA champions league tournament, the average number of goals scored in a match is 2.3. 

**What is the probability that the number of goals scored in a match is 5?**

*Hint: Use the exp() in the math library*

And to calculate the factorial, you can use the following code:

```python
import math
math.factorial(n) # n is the number you want to find the factorial of
```

````python
# your code here 
import math

p_X5 = (math.exp(-2.3)*((2.3)**5 ))/ math.factorial(5)

print(f'the probability that the number of goals scored in a match is 5: {p_X5:.3f}')
````

**Draw a poisson probability distribution plot for the number of goals to be between 0 to 10.**

````python
# your code here
# Please label the axes and give a title to the plot 
x=range(0,11)
y = []
for i in x:
    prob = (math.exp(-2.3)*((2.3)**i))/ math.factorial(i)
    y.append(prob)
plt.bar(x,y, color = 'tan')

plt.rcParams["font.family"]= "Arial"
plt.tick_params(axis= 'both', direction = 'in')
plt.xlabel('number of goals', fontsize = 12)
plt.ylabel('probability /a.u.', fontsize = 12)
plt.title('poisson probability distribution plot for the number of goals to be between 0 to 10', fontsize = 12)
````
